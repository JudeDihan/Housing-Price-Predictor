Now let's update the transform_features() function and see how our RMSE scores are

```python
def transform_features(dframe):
    num_missing = (dframe.isnull().sum())/dframe.shape[0]*100
    drop_cols = num_missing[num_missing > 5]
    drop_cols_names = drop_cols.index
    dframe = dframe.drop(drop_cols_names, axis=1)
    
    
    text_cols = dframe.select_dtypes(include='object')
    missing_text_cols = text_cols.isnull().sum().sort_values(ascending=False)
    missing_text_cols = missing_text_cols[missing_text_cols > 0].index
    dframe = dframe.drop(missing_text_cols, axis=1)
    
    numerical_cols = dframe.select_dtypes(include=['float','integer'])
    missing_cols = numerical_cols.isnull().sum().sort_values(ascending=False)
    missing_cols = missing_cols[missing_cols>0]
    mode_missing_cols = dframe[missing_cols.index].mode().to_dict(orient='records')[0]
    dframe = dframe.fillna(mode_missing_cols)
    
    years_sold = dframe['Yr Sold'] - dframe['Year Built']
    years_since_remod = dframe['Yr Sold'] - dframe['Year Remod/Add']
    dframe['Years Before Sale'] = years_sold
    dframe['Years Since Remod'] = years_since_remod
    dframe = dframe.drop([1702, 2180, 2181], axis=0)
    dframe = dframe.drop(["Year Built", "Year Remod/Add"], axis = 1)
    
    dframe = dframe.drop(["PID", "Order"], axis=1)
    dframe = dframe.drop(["Mo Sold", "Sale Condition", "Sale Type", "Yr Sold"], axis=1)
    return dframe

def select_features(dframe):
    return dframe[['Gr Liv Area', 'SalePrice']]


def train_and_test(dframe):
    train = dframe.iloc[:1460]
    test = dframe.iloc[1460:]
    
    # Let's select only the numerical columns from the dataset
    numericals_train = train.select_dtypes(include=['float','integer'])
    numericals_test = test.select_dtypes(include=['float','integer'])
    
    # Make sure we drop the target column (SalePrice) before training the model
    numericals_cols = numericals_train.columns.drop('SalePrice')
    
    regressor = LinearRegression()
    regressor.fit(train[numericals_cols], train['SalePrice'])
    predictions = regressor.predict(test[numericals_cols])
    
    #Calculations of RMSE
    mse = mean_squared_error(test['SalePrice'], predictions)
    rmse = np.sqrt(mse)
    
    return rmse
    
transform_df = transform_features(dataset)    
filtered = select_features(transform_df)
rmse = train_and_test(filtered)
print(rmse)
```

# 4. Train and Test

Now let's include the above in the function and also train and test 

```python
def transform_features(dframe):
    num_missing = (dframe.isnull().sum())/dframe.shape[0]*100
    drop_cols = num_missing[num_missing > 5]
    drop_cols_names = drop_cols.index
    dframe = dframe.drop(drop_cols_names, axis=1)
    
    
    text_cols = dframe.select_dtypes(include='object')
    missing_text_cols = text_cols.isnull().sum().sort_values(ascending=False)
    missing_text_cols = missing_text_cols[missing_text_cols > 0].index
    dframe = dframe.drop(missing_text_cols, axis=1)
    
    numerical_cols = dframe.select_dtypes(include=['float','integer'])
    missing_cols = numerical_cols.isnull().sum().sort_values(ascending=False)
    missing_cols = missing_cols[missing_cols>0]
    mode_missing_cols = dframe[missing_cols.index].mode().to_dict(orient='records')[0]
    dframe = dframe.fillna(mode_missing_cols)
    
    years_sold = dframe['Yr Sold'] - dframe['Year Built']
    years_since_remod = dframe['Yr Sold'] - dframe['Year Remod/Add']
    dframe['Years Before Sale'] = years_sold
    dframe['Years Since Remod'] = years_since_remod
    dframe = dframe.drop([1702, 2180, 2181], axis=0)
    dframe = dframe.drop(["Year Built", "Year Remod/Add"], axis = 1)
    
    dframe = dframe.drop(["PID", "Order"], axis=1)
    dframe = dframe.drop(["Mo Sold", "Sale Condition", "Sale Type", "Yr Sold"], axis=1)
    return dframe


def select_features(df, coeff_threshold=0.4, uniq_threshold=10):
    numerical_df = df.select_dtypes(include=['int', 'float'])
    abs_corr_coeffs = numerical_df.corr()['SalePrice'].abs().sort_values()
    df = df.drop(abs_corr_coeffs[abs_corr_coeffs < coeff_threshold].index, axis=1)
    
    nominal_features = ["PID", "MS SubClass", "MS Zoning", "Street", "Alley", "Land Contour", "Lot Config", "Neighborhood", 
                    "Condition 1", "Condition 2", "Bldg Type", "House Style", "Roof Style", "Roof Matl", "Exterior 1st", 
                    "Exterior 2nd", "Mas Vnr Type", "Foundation", "Heating", "Central Air", "Garage Type", 
                    "Misc Feature", "Sale Type", "Sale Condition"]
    
    transform_cat_cols = []
    for col in nominal_features:
        if col in df.columns:
            transform_cat_cols.append(col)

    uniqueness_counts = df[transform_cat_cols].apply(lambda col: len(col.value_counts())).sort_values()
    drop_nonuniq_cols = uniqueness_counts[uniqueness_counts > 10].index
    df = df.drop(drop_nonuniq_cols, axis=1)
    
    text_cols = df.select_dtypes(include=['object'])
    for col in text_cols:
        df[col] = df[col].astype('category')
    df = pd.concat([df, pd.get_dummies(df.select_dtypes(include=['category']))], axis=1).drop(text_cols,axis=1)
    
    return df


def train_and_test(df, k=0):
    numeric_df = df.select_dtypes(include=['integer', 'float'])
    features = numeric_df.columns.drop("SalePrice")
    lr = LinearRegression()
    
    if k == 0:
        train = df[:1460]
        test = df[1460:]

        lr.fit(train[features], train["SalePrice"])
        predictions = lr.predict(test[features])
        mse = mean_squared_error(test["SalePrice"], predictions)
        rmse = np.sqrt(mse)

        return rmse
    
    if k == 1:
        # Randomize *all* rows (frac=1) from `df` and return
        shuffled_df = df.sample(frac=1, )
        train = df[:1460]
        test = df[1460:]
        
        lr.fit(train[features], train["SalePrice"])
        predictions_one = lr.predict(test[features])        
        
        mse_one = mean_squared_error(test["SalePrice"], predictions_one)
        rmse_one = np.sqrt(mse_one)
        
        lr.fit(test[features], test["SalePrice"])
        predictions_two = lr.predict(train[features])        
       
        mse_two = mean_squared_error(train["SalePrice"], predictions_two)
        rmse_two = np.sqrt(mse_two)
        
        avg_rmse = np.mean([rmse_one, rmse_two])
        print(rmse_one)
        print(rmse_two)
        return avg_rmse
    else:
        kf = KFold(n_splits=k, shuffle=True)
        rmse_values = []
        for train_index, test_index, in kf.split(df):
            train = df.iloc[train_index]
            test = df.iloc[test_index]
            lr.fit(train[features], train["SalePrice"])
            predictions = lr.predict(test[features])
            mse = mean_squared_error(test["SalePrice"], predictions)
            rmse = np.sqrt(mse)
            rmse_values.append(rmse)
        print(rmse_values)
        avg_rmse = np.mean(rmse_values)
        return avg_rmse

df = pd.read_csv("AmesHousing.tsv", delimiter="\t")
transform_df = transform_features(df)
filtered_df = select_features(transform_df)
rmse = train_and_test(filtered_df, k=4)

rmse
```
